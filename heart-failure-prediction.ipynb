```python
# Getting data ready
import pandas as pd

heart_failure = pd.read_csv("./data/heart_failure_clinical_records.csv")
heart_failure.head(3)
```

```python
X=heart_failure.drop(labels="DEATH_EVENT", axis=1)
Y=heart_failure["DEATH_EVENT"]
```

```python
# Splitting data
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.8)

X_train
```

```python
# Create a model
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()

model.get_params()
```

```python
# Train/Fit the model

model.fit(X_train, Y_train)

model.score(X_train, Y_train)
```

```python
# Making predictions with the model
model.predict(X_train)
```

```python
# Evaluate the model
model.score(X_train, Y_train), model.score(X_test, Y_test)
```

```python
# Improve the model (changing n_estimators)
for i in range (10, 100, 10):
    model = RandomForestClassifier(i)
    model.fit(X_train, Y_train)
    print(f"Model with {i} n_estimators hits {model.score(X_test, Y_test)} accuracy on test set")
```

```python
# Save model with the best score
model = RandomForestClassifier(50)
model.fit(X_train, Y_train)

import pickle
pickle.dump(model, open("Heart_Failure_Prediction.pkl", "wb"))
```
